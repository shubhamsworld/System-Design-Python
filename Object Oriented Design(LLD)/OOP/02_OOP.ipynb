

# Object-Oriented Programming (OOP) Fundamentals

## Introduction

Object-Oriented Programming (OOP) is a programming paradigm used to model real-world systems in software.

The primary goal of OOP is to reduce complexity by representing a software system as a collection of interacting objects.

Instead of thinking in terms of functions and data separately, OOP combines data and behavior into a single unit called an object.

---

# The Foundation of OOP

Object-Oriented Programming is built upon a fundamental principle called:

## Abstraction

Abstraction is the foundation on which the entire Object-Oriented Design is based.

The other concepts of OOP exist mainly to help implement abstraction effectively.

---

OOP is commonly explained using:

## Three Pillars

## One Fundamental Principle

1. Abstraction

```text
                 Abstraction
                      ▲
                      │
      ┌───────────────┼───────────────┐
      │               │               │
Encapsulation    Inheritance    Polymorphism
```

# What is a Principle?

A principle is a fundamental idea or value system that guides the design of software.

Examples:

* Abstraction
* Separation of Concerns
* Single Responsibility

A principle tells us:

```text
What should be achieved?
```

```text
How should it be achieved?
```

# Abstraction

## Definition

Abstraction means representing a complex system in terms of simple ideas while hiding unnecessary implementation details.

Instead of focusing on how something works internally, we focus on what it does.

---

## Understanding Abstraction Through Real Life

Consider a car.

As a driver, you know:

* Steering wheel turns the car
* Brake stops the car
* Accelerator increases speed

You do not need to know:

* Fuel injection mechanism
* Combustion process
* Gear synchronization
* Engine internals

You simply use the exposed functionality.

This is abstraction.

---

## Abstraction in Software

A complex software system may contain:

* Databases
* APIs
* Business Logic
* Authentication
* Caching
* Logging

A user of the system does not need to understand all these details.

Instead, the system is represented using simple ideas.

Example:

```text
Customer
Order
Payment
Invoice
Product
```

Each of these ideas hides internal complexity.

---

## Abstraction in Object-Oriented Design

In OOP, we represent a system using entities.

Each entity has:

### Data (Attributes)

Information related to the entity.

Example:

```text
Customer
 ├── Name
 ├── Email
 └── Phone Number
```

---

### Behavior (Methods)

Actions that the entity can perform.

Example:

```text
Customer
 ├── Register()
 ├── Login()
 └── UpdateProfile()
```

---

## Key Idea

Abstraction means:

> Representing a complex software system in terms of meaningful ideas, where each idea contains data and behavior, while hiding unnecessary implementation details.

---

## Why Abstraction is Important

Abstraction helps us:

### Reduce Complexity

Instead of understanding thousands of lines of code, we understand a few concepts.

---

### Improve Readability

Systems become easier to understand.

---

### Improve Maintainability

Internal implementation can change without affecting users of the object.

---

### Improve Reusability

Well-designed abstractions can be reused across projects.

---

# How Abstraction is Achieved

Programming languages cannot implement abstraction automatically.

To implement abstraction, OOP provides several mechanisms:

1. Encapsulation
2. Inheritance
3. Polymorphism

These are known as the pillars of OOP.

---

# Encapsulation

## Meaning of Encapsulation

The word encapsulation comes from the word "capsule".

Think about a medicine capsule.

A capsule:

* Holds medicine together
* Protects the medicine from the external environment
* Hides unnecessary details
* Provides a simple interface to the user

The user only consumes the capsule.

The user does not care how the medicine is packaged internally.

---

## Encapsulation in OOP

Encapsulation means:

> Bundling data and behavior together into a single unit and controlling access to the internal state.

In OOP, that unit is called an Object.

---

## Example

Without encapsulation:

```python
account_balance = 1000
```

Anyone can directly modify:

```python
account_balance = -5000
```

This may create invalid states.

---

With encapsulation:

```python
account.deposit(500)
account.withdraw(200)
```

The object itself controls how its data can be modified.

---

## Benefits of Encapsulation

### 1. Data Protection

Internal data is protected from improper access.

---

### 2. Better Control

Objects decide what operations are allowed.

---

### 3. Reduced Complexity

Clients only see what is necessary.

---

### 4. Easier Maintenance

Internal implementation can change without affecting users.

---

## Access Modifiers

Encapsulation is often implemented using access modifiers.
| Modifier  | Syntax   | Meaning                                  |
| --------- | -------- | ---------------------------------------- |
| Public    | `name`   | Accessible from anywhere                 |
| Protected | `_name`  | Intended for internal use and subclasses |
| Private   | `__name` | Name mangling applied                    |


These determine what can be accessed from outside the class.

# Terms Used in OOP

Before learning further concepts, we need to understand two important terms.

---

## Class

A class is a blueprint or template for creating objects.

It describes:

* What data an object will contain
* What behavior an object will have

A class itself does not represent a real entity.

It is only a design.

Example:

```python
class Car:
    pass
```

Think of a class as an architectural blueprint of a house.

The blueprint describes the house but is not the actual house.

---

## Object

An object is a real instance of a class.

Objects are created using classes.

Example:

```python
car1 = Car()
car2 = Car()
```

Here:

* `car1` is an object
* `car2` is another object

Both are created from the same blueprint.

---

## Class vs Object

| Class                         | Object                    |
| ----------------------------- | ------------------------- |
| Blueprint                     | Real instance             |
| Logical concept               | Physical entity in memory |
| Does not store actual data    | Stores actual data        |
| Used to create objects        | Created from a class      |
| Does not occupy object memory | Occupies memory           |

## Relationship Between Class, Encapsulation and Abstraction

Class is the primary mechanism used to implement Encapsulation.

Encapsulation bundles data and behavior together and controls access to internal details.

By hiding unnecessary implementation details and exposing only relevant behavior, Encapsulation helps achieve Abstraction.


Abstraction is the goal.
Encapsulation is one of the mechanisms used to achieve that goal.
Classes and Objects are the language constructs that make Encapsulation possible.

```text
Abstraction
    ↑
Achieved using
    ↑
Encapsulation
    ↑
Implemented using
    ↑
Classes and Objects
```

Abstraction = Hiding details
Encapsulation = Hiding data
