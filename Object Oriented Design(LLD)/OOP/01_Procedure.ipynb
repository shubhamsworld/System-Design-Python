
# Procedural Programming

## What is a Procedure?

A procedure is a set of instructions that performs a specific task.

A procedure:

* Can take zero or more inputs (parameters)
* Performs some actions
* May or may not return a value

In modern programming languages, procedures are usually called **functions** or **methods**.

Example:

```python
def greet(name):
    print(f"Hello {name}")
```

Here, `greet()` is a procedure/function.

---

## Characteristics of Procedural Programming

In procedural programming:

* The codebase is divided into multiple functions/procedures.
* Each procedure performs a specific task.
* Procedures can call other procedures.
* Program execution usually starts from a special function such as `main()`.

Example:

```python
def read_data():
    pass

def process_data():
    pass

def save_data():
    pass

def main():
    read_data()
    process_data()
    save_data()
```

---

# The Role of Data

Every software system deals with:

* Entities (Objects in the real world)
* Data related to those entities
* Actions performed on those entities

Examples:

| Entity       | Data                    |
| ------------ | ----------------------- |
| Student      | Name, Roll Number       |
| Employee     | Name, Salary            |
| Car          | Model, Speed            |
| Bank Account | Balance, Account Number |

---

# How Procedural Programming Represents the Real World

In procedural programming:

* Data and behavior are separate.
* Entities are represented as variables, dictionaries, structures, or records.
* Actions are represented as functions.

Example:

```python
employee = {
    "name": "John",
    "salary": 50000
}

def increase_salary(employee, amount):
    employee["salary"] += amount
```

Here:

* `employee` contains data.
* `increase_salary()` contains behavior.
* Data and behavior are separate.

---

# Real World vs Procedural Programming

## Real World

Entities perform actions.

Examples:

```text
A bank account withdraws money.
A car starts.
An employee receives a salary increment.
```

The entity owns its behavior.

---

## Procedural Programming

Actions are performed on entities.

Examples:

```python
withdraw(account)
start_car(car)
increase_salary(employee)
```

The behavior exists outside the entity.

The entity itself is only data.

---

# The Main Limitation

As software grows, many functions start operating on the same data.

Example:

```python
create_employee()
update_employee()
delete_employee()
calculate_salary()
print_employee()
validate_employee()
```

All these functions depend on the same employee data.

If the employee structure changes, many functions must also change.

---

# Problems with Procedural Programming

## 1. Difficult to Understand in Large Systems

When a project becomes large:

* Functions are spread across many files.
* Understanding the flow becomes difficult.
* It becomes hard to identify where a particular action is happening.

---

## 2. Difficult to Maintain

A change in data structure may require updates in multiple functions.

Example:

```python
employee = {
    "name": "",
    "salary": 0
}
```

Later:

```python
employee = {
    "name": "",
    "salary": 0,
    "department": ""
}
```

Many functions may need modification.

---

## 3. Difficult to Debug

When a bug occurs:

* Multiple functions may be involved.
* Data can be modified from many different places.
* Tracing the root cause becomes harder.

---

## 4. High Dependency Between Functions

Functions often depend on each other.

Example:

```text
Function A
    ↓
Function B
    ↓
Function C
    ↓
Function D
```

A change in one function may affect several others.

This increases coupling.

---

## 5. Spaghetti Code

As the system grows:

```text
A → B → C
↓
D → E
↓
F → G → H
```

Function calls become complex and difficult to track.

This situation is commonly called **Spaghetti Code**.

---

# How Object-Oriented Programming Solves This

Object-Oriented Programming combines:

```text
Data + Behavior
```

into a single unit called an **Object**.

Instead of:

```python
withdraw(account, amount)
```

we write:

```python
account.withdraw(amount)
```

The object now controls:

* Its own data
* Its own behavior
* How its data can be modified

This makes software:

* Easier to understand
* Easier to maintain
* Easier to extend
* Easier to debug

---

# Key Takeaway

### Procedural Programming

```text
Data and behavior are separate.
Functions operate on data.
```

### Object-Oriented Programming

```text
Data and behavior are bundled together.
Objects manage their own state and behavior.
```

This limitation of procedural programming is one of the main reasons why Object-Oriented Programming became popular for building large and complex software systems.
